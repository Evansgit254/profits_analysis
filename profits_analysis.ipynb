```python
#Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Load and analyze data
#Upload the provided data file on Colab and load the data using Pandas
data = pd.read_csv('dataset_assignment 2.csv')
data.head()
```

```python
#Handle Missing Values
data.isnull().sum()
```

From the above output,we have missing values in the following columns:UnitCost,Sales and Profit with each having 100 missing values as shown in the above output.

```python
#plots before handling the missing values in the data
#Apply a moving average technique to smooth out the data and observe any patterns or trends
#UnitCost feature plot
window_size = 100
moving_average = data['UnitCost'].rolling(window_size).mean()
plt.plot(data['UnitCost'], 'b', moving_average, 'r')
```

```python
#Sales feature plot
window_size = 100
moving_average = data['Sales'].rolling(window_size).mean()
plt.plot(data['Sales'], 'b', moving_average, 'r')
```

```python
#Profit feature plot
window_size = 100
moving_average = data['Profit'].rolling(window_size).mean()
plt.plot(data['Profit'], 'b', moving_average, 'r')
```

```python
#Handle Missing Values
#Fill missing values with interpolation technique
data_interpolated = data.interpolate()
```

```python
data_interpolated.isnull().sum()
```

From the output above, we can see that the missing values in UnitCost, Sales and Profit columns have been handled there are no values missing.

```python
#plots after handling missing values in the data for each feature.
#plot for UnitCost feature
window_size = 100
moving_average = data_interpolated['UnitCost'].rolling(window_size).mean()
plt.plot(data_interpolated['UnitCost'], 'b', moving_average, 'r')
```

```python
#plot for Sales feature
window_size = 100
moving_average = data_interpolated['Sales'].rolling(window_size).mean()
plt.plot(data_interpolated['Sales'], 'b', moving_average, 'r')
```

```python
#plot for Profit feature
window_size = 100 # Adjust the window size as desired
moving_average = data_interpolated['Profit'].rolling(window_size).mean()
plt.plot(data_interpolated['Profit'], 'b', moving_average, 'r')
```

From the two scenarios the "before" and "after" handling missing values,the output plots of "after" present a smooth red moving average in each feature showing the trend of the data plots whereas the plots of "before" do not ouput a moving average hence does not indicate any trends and patterns in the data.

```python
#Identify Outliers
#Identify any outliers in the data and discuss their potential impact on the central tendencies
#UnitCost Feature plot
ax = data_interpolated['UnitCost'].plot(kind='hist',
bins=20,
title='',log=True)
ax.set_xlabel('UnitCost')
```

We have outliers on the right side of the histogram that's from around 15 to 30 values of the UniCost along the x_axis.

```python
#Identify Outliers
#Identify any outliers in the data and discuss their potential impact on the central tendencies.
#Sales feature plot
ax = data_interpolated['Sales'].plot(kind='hist',
bins=20,
title='',log=True)
ax.set_xlabel('Sales')
```

```python
#Identify Outliers
#Identify any outliers in the data and discuss their potential impact on the central tendencies.
#Profit feature plot
ax = data_interpolated['Profit'].plot(kind='hist',
bins=20,
title='',log=True)
ax.set_xlabel('Profit')
```

```python
#Calculate Correlation
#Calculate the correlation coefficient between two variables in the data set and interpret the strength and direction of the correlation.
#correlation between unitcost and sales
correlation = data_interpolated ['UnitCost'].corr(data['Sales'])
print(correlation)
```

From the above ouput, it indicates that there is positive correlation between UnitCost and sales.
